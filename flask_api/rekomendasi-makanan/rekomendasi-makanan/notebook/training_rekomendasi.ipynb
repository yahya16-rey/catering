```python
import pandas as pd
import joblib

from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv('../dataset/menu.csv')

df.head()
```

```python
df['fitur'] = (
    df['nama_menu'] + ' ' +
    df['kategori'] + ' ' +
    df['deskripsi']
)

df[['nama_menu', 'fitur']].head()
```

```python
factory = StemmerFactory()
stemmer = factory.create_stemmer()

def preprocessing(text):

    text = text.lower()

    text = stemmer.stem(text)

    return text
```

```python
df['fitur'] = df['fitur'].apply(preprocessing)

df[['nama_menu', 'fitur']].head()
```

```python
tfidf = TfidfVectorizer()

tfidf_matrix = tfidf.fit_transform(df['fitur'])

print(tfidf_matrix.shape)
```

```python
cosine_sim = cosine_similarity(tfidf_matrix)

print(cosine_sim)
```

```python
indices = pd.Series(df.index, index=df['nama_menu']).drop_duplicates()

indices
```

```python
def recommend_menu(nama_menu, top_n=5):

    idx = indices[nama_menu]

    sim_scores = list(enumerate(cosine_sim[idx]))

    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    sim_scores = sim_scores[1:top_n+1]

    menu_indices = [i[0] for i in sim_scores]

    hasil = df[
        ['nama_menu', 'kategori', 'harga', 'gambar']
    ].iloc[menu_indices]

    return hasil
```

```python
recommend_menu('Nasi Ayam Geprek')
```

```python
joblib.dump(tfidf, '../model/tfidf_vectorizer.pkl')

joblib.dump(cosine_sim, '../model/cosine_similarity.pkl')

joblib.dump(indices, '../model/menu_indices.pkl')

print("Model berhasil disimpan")
```
